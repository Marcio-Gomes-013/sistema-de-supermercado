# Descrição do Sistema



 - Sistema de Supermecado. Nesse sistema vamos interagir de duas formas, a primeira vamos desenha a tela de cadastro dos produtos, na segunda vamos criar a tela de interação do cliente, ou seja, vamos cria a tela de compra.

 1 - Tela de Cadastro de Produtos
    - Cadastro de Fornecedor
      - Nome
      - Categoria Produto
      - Produto
      - Tipo(CX, PCT, KG)
      - Quantidade
      - Valor
    - Cadastro de Categorias
      - Nome Categoria
    - Cadastro de Produtos
      - Nome Produto
      - Tipo Produto
      - Quantidade
      - Valor Unitario
    - Cadastro de Sessão
      - Nome da Sessão

    - Lista Fornecedor, Categorias, Produtos e Sessão
    
 2 - Tela de Compra do Cliente
    - Lista de Produtos
    - Adicionar Produto ao Carrinho
    - Remover Produto do Carrinho
    - Subtotal da Compra
    - Finalizar Compra
    - Lista Produtos quantidade e Valor

# Bibliotecas

```
import os
import csv
from csv import DictReader, DictWriter
```

# Tela Administrador

```
# Cadastros Fornecedor, Categorias, Produtos, Sessão

def cadastroFornecedor(arquivo):
  filename = f"{arquivo}.csv"
  if os.path.exists(filename) == False:
    with open(filename, "w") as arquivo:
      cabecalho = ['Fornecedor', 'Categoria', 'Produto', 'Tipo', 'Quantidade', 'Valor']
      escrever = DictWriter(arquivo, fieldnames=cabecalho)
      escrever.writeheader()
      nome = None
      while nome != 'sair':
          nome = input('Informe o nome do fornecedor: ')
          if nome != 'sair':
              categoria   = input('Informe a categoia do produto: ')
              produto     = input('Informe o produto: ')
              tipo        = input('Informe o tipo do produto: ')
              quantidade  = int(input('Informe a quantidade do produto: '))
              valor       = float(input('Informe o valor do produto: '))
              escrever.writerow({"Fornecedor": nome, 
                                    "Categoria": categoria, 
                                    "Produto": produto,
                                    "Tipo":tipo,
                                    "Quantidade": quantidade,
                                    "Valor": valor
                                    })
  elif os.path.exists(filename) == True:
    with open(filename, 'a') as arquivo:
      escrever = csv.writer(arquivo, delimiter=',')
      nome = None
      while nome != 'sair':
        nome = input('Informe o nome do fornecedor: ')
          if nome != 'sair':
            categoria   = input('Informe a categoia do produto: ')
            produto     = input('Informe o produto: ')
            tipo        = input('Informe o tipo do produto: ')
            quantidade  = int(input('Informe a quantidade do produto: '))
            valor       = float(input('Informe o valor do produto: '))
            escrever.writerow([nome, categoria, produto, tipo, quantidade, valor])


```

```

def cadastro(tipoCadastro, colunas=[]):
  filename = f"{tipoCadastro}.csv"
  if os.path.exists(filename) == False:
    print(f"Arquivo Não Existe {filename}")
    with open(filename, 'w') as arquivo:
      escreveArquivo = DictWriter(arquivo, fieldnames=colunas)
      escreveArquivo.writeheader()
      colunas[0] = None
      while colunas[0] != 'sair':
        for itens in colunas:
          dados = input(f"Digite o {itens} ")
          if colunas[0] != 'sair':
            print(dados)

  else:
    print(f"Arquivo Existe {filename}")
    with open(filename, 'w') as arquivo:
      escreveArquivo = DictWriter(arquivo, fieldnames=colunas)
      escreveArquivo.writeheader()

def leitura(fonteDados):
  filename = f"{fonteDados}.csv"
  with open(filename, 'r') as arquivo:
    lerColunas = csv.DictReader(arquivo)
    colunasArquivo = lerColunas.fieldnames
    print(colunasArquivo)


cadastro('fornecedor',['Nome', 'Categoria Produto', 'Produto', 'Quantidade', 'Valor'])
```
